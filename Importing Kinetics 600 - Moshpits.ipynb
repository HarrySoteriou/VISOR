# Downloading from youtube

```python
from pytube import YouTube
import os
from __future__ import unicode_literals
import youtube_dl
import urllib
import shutil
import ffmpeg
import imageio
from moviepy.editor import *
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import glob
```

```python
#download function
def downloadYouTube(videourl, path):
    yt = YouTube(videourl)
    yt = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
    if not os.path.exists(path):
        os.makedirs(path)
    yt.download(path)

#downloadYouTube('https://www.youtube.com/watch?v=FN4bzVdySAA', './Youtube Videos')
```

```python
training_moshpit_list = [training_data[i] for i in training_data.keys() if training_data[i]['annotations']['label'] == 'mosh pit dancing']
training_moshpit_list = training_moshpit_list[440:]
```

```python
# Get the latest file in the folder
def get_latest_file():
    files_path = os.path.join('./Youtube Videos', '*')
    files = sorted(glob.iglob(files_path), key=os.path.getctime, reverse=True) 
    #return("'" + files[0] + "'")
    return(files[0])
```

```python
# Final algo
def cut_all_videos():
    """ get a list of all video names in the folder, iterate over them and convert them according to their start and ending times"""
    urls = [video['url'] for video in training_moshpit_list] #create list of urls
    i = 441
    for video in training_moshpit_list:
        downloadYouTube(video['url'], './Youtube Videos')
        ffmpeg_extract_subclip(get_latest_file(), int(video['annotations']['segment'][0]), int(video['annotations']['segment'][1]), targetname=('Moshpit Train Videos/'+ str(i) + '.mp4'))
        i += 1

cut_all_videos()
```
